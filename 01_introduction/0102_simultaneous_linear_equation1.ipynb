# 連立一次方程式

まず，線形代数への入口として連立一次方程式について確認しましょう．
私たちと線形代数との出会いは連立一次方程式の問題と言っていいでしょう．


## 初歩的な問題

連立一次方程式の問題設定となる初歩的な文章問題を提示します．



#### 【例題1】

果物の詰め合わせセットが2種類あります．一つ目の詰め合わせセットにはリンゴが2個とオレンジが8個入っていて560円です．もう一つの詰め合わせセットにはリンゴが5個とオレンジが3個入って550円です．
リンゴとオレンジの単価はそれぞれ幾らでしょうか．

#### 定式化

問題の文章を読み解くと2つの方程式が導けます．<font color=green>リンゴの単価を $x$ 円</font>とし，<font color=green>オレンジの単価を $y$ 円</font>とします．まず，一つ目の詰め合わせセットの価格についての文章を数式で表します．


> <font color=green>リンゴが2個とオレンジが8個入っていて560円</font>

> $2x + 8y = 560$

二つ目の詰め合わせセットの価格についての文章も数式化します．

> <font color=green>リンゴが5個とオレンジが3個入って550円</font>

> $5x + 3y = 550$

文章問題から2つの方程式が求まりました．
このように，問題を数式に変換してしまえば，その後は純粋に数学の問題として扱うことができます．
よって，例題1は次に純粋な数学の問題に書き換えることができます．

#### 【例題1'】

次の連立一次方程式の解となる$x$と$y$求めよ．

$\left\{\begin{array}{lc}
2x + 8y = 560 & \cdots (1)\\  
5x + 3y = 550 & \cdots (2)\\
\end{array} \right.$

ただし制約条件として，$x,y \in \mathbb{Z}, x\ge0, y\ge0$を満たすものとする．
この制約条件は，元の問題がリンゴとオレンジの単価を求めることに付随しています．

ここで，連立一次方程式の問題に表れる数学用語について確認しておきましょう．

まず，リンゴとオレンジの単価をそれぞれ$x$円と$y$円という文字で置き換えました．この$x$と$y$の具体的な値は未だ分からないので**未知数**と言います．また，未知数は方程式の中では**変数**と呼ばれます．
**方程式**とは一つ以上の変数を含む等式で，変数の値によって等号が成り立つ場合と成り立たない場合がありえます．

**一次方程式**とは変数の多項式で構成される方程式のうち定数項および変数の一次項しか含まない式です．
1変数の一次方程式は，$x$を唯一の変数としたとき$a$と$b$を定数として$ax = b$の形式で表されます．
一次方程式でない例としては，$x^2 - 5x + 6 = 0$があります．これは$x$の2乗の項があるので二次方程式になります．

**連立方程式**とは，複数の方程式が同時に成り立つことを想定します．
例題1'においては，一次方程式(1)が成立し，かつ，一次方程式(2)が成立する場合を想定します．

それでは，連立一次方程式の【例題1'】を初等的な解法によって解いてみましょう．

#### 【例題1解答】

【例題1'】を単純に解く方法について確認しましょう．
連立方程式を解くための方針は，一つの方程式における変数の個数を減らすことです．すなわち，方程式の等号関係を維持しながら$ax = b$という方程式を導くことができれば$x = b/a$と求められます．そのために式の変形を行います．

まず式(1)の両辺を2で割ります．

> $x + 4y = 280 \ \ \ \  \cdots (3)$

次に式(2)から式(3)を5倍にした式を引きます．その結果を整理すると変数$y$についての一次方程式となり，$y$の値が求まります．

> $ \begin{eqnarray}
(5x + 3y) - 5(x + 4y) &=& 550 - 5\times280 \\
(5-5)x + (3-20)y &=& 550 - 1400 \\
-17y &=& -850 \\
y &=& (-850)\div(-17) \\
\end{eqnarray} $

> $y = 50 \ \ \ \ \ \cdots (4)$

そして，式(4)を式(3)に代入して，$x$の値を求めます．

> $ \begin{eqnarray}
x + 4\times50 &=& 280 \\
x &=& 280 - 200 \\
\end{eqnarray} $

> $x = 80 \ \ \ \ \ \cdots (5)$

以上で【例題1'】が解けました．この結果を元の文章題の題意に戻すと，リンゴの単価は80円でオレンジの単価が50円という解答になります．

## 練習問題

例題1と同様の問題について，問題の定式化を行い，連立方程式を解いてみましょう．
次の問題は変数が3個ある連立方程式の問題となります．

#### 【例題2】

3種類のコーヒー豆A，B，Cを購入しました．100g当たりの単価はそれぞれ800円，400円，100円です．
A，B，C合わせて700g購入しましたが，A，Bの合計量はCの量より100g少なかったです．
合計金額は2,000円でした．
A，B，Cそれぞれ何g購入したか求めなさい．

#### 問題の定式化

まず，変数を導入する．

- $x$ ：コーヒー豆Aの購入量(g)
- $y$ ：コーヒー豆Bの購入量(g)
- $z$ ：コーヒー豆Cの購入量(g)

すると，A，B，C合わせて700g購入したので，

> $x + y + z = 700$

が成立します．
次に，コーヒー豆A，B，Cの100g当たりの単価が800円，400円，100円なので1g当たりの単価は8円，4円，1円となり，合計金額が2,000円なので，

> $8x + 4y + z = 2000$

となります．
さらに，A，Bの合計量はCの量より100g少なかったので，

> $x + y = z - 100$

が成り立ちます．
これをまとめると，次の連立一次方程式になります．

#### 【例題2'】

次の連立一次方程式の解となる$x, y, z$を求めよ．

$\left\{\begin{array}{lc}
 x +  y + z = 700  & \cdots (1) \\ 
8x + 4y + z = 2000 & \cdots (2) \\ 
 x +  y - z = -100 & \cdots (3) \\
\end{array} \right.$

ただし，制約条件として$x,y,z \in \mathbb{R}, x\ge0, y\ge0, z\ge0$を満たすものとする．この条件は求めるものがコーヒー豆の量なので負の値を取らないということです．

#### 【例題2解答】

式(1)を使って，式(2)および式(3)から変数$x$を消去します．
まずは式(2)から式(1)の8倍を引きます．

> $(8x + 4y + x) - (8x + 8y + 8z) = 2000 - 8\times700$

> $-4y - 7z = -3600 \ \ \ \ \cdots (4)$

同様に式(3)から式(1)を引きます．その後，両辺を-2で割ります．

> $\begin{eqnarray}
(x + y - z) - (x + y + z) &=& -100 - 700 \\
-2z &=& -800 \\
\end{eqnarray}$

> $z = 400 \ \ \ \ \cdots (5)$

これで$z$の値が求まりましたので，式(5)を式(4)に代入して$y$の値を求めます．

> $\begin{eqnarray}
-4y - 7\times400 &=& -4000 \\
-4y &=& -3600 + 2800 \\
y &=& -800\div(-4) \\
\end{eqnarray}$

> $y = 200 \ \ \ \ \cdots (6)$

これで$y$の値も求まりました．
最後に式(1)に式(5)および式(6)を代入して$x$の値を求めます．

> $\begin{eqnarray}
x + 200 + 400 &=& 700 \\
x &=& 700 - 200 - 400 \\
\end{eqnarray}$

> $x = 100 \ \ \ \ \cdots (7)$

となり$x$の値も求まりました．
文章題に戻すと，コーヒー豆A，B，Cをそれぞれ100g，200g，400g購入したことになります．

## 数学の真髄

ここまでの問題を解くことによって，典型的な線形代数の問題である連立一次方程式について見てきました．
ここでの解法では，変数を含む数式を地道に計算して解を求めるという計算問題に過ぎません．
これだけですと線形代数学という数学の一分野を名乗るほどのものではないように思えます．

私たちは，連立方程式の解法から始めて，線形代数の数学的構造について学習していきます．
