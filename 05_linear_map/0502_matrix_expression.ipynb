# 線形写像の表現行列

線形写像の概念について学習しましたが，ベクトル空間における線形写像は行列によって表すことができます．
このことで，線形写像を研究するためには行列を詳しく調べれば良いことが分かります．

## 表現行列定理

$\bf V$を$n$次元のベクトル空間，$\bf W$を$m$次元のベクトル空間とし，それぞれのスカラー$K$は同じものとします．
$\bf V$の基底を ${\bf v}_1,\cdots,{\bf v}_n$，$\bf W$の基底を ${\bf w}_1,\cdots,{\bf w}_m$ とします．
このとき，線形写像 $f:{\bf V} \rightarrow {\bf W}$に対して，

$ \left(\begin{array}{c} f({\bf v}_1) & \cdots & f({\bf v}_n) \end{array}\right) = 
\left(\begin{array}{c} {\bf w}_1 & \cdots & {\bf w}_m \end{array}\right) {\bf M}_f $

となる$m$行$n$列の行列${\bf M}_f$が存在する．
この行列を線形写像 $f$ の<font color=blue>表現行列</font>と言います．

## 表現行列の例

これまで見てきた幾つかの線形写像について，その表現行列を確認します．

*****
## 射影の表現行列

3次元の実数ベクトル空間$\mathbb{R}^3$から2次元の実数ベクトル空間$\mathbb{R}^2$への射影

> $P_{12}: \left(\begin{array}{c} x_1 \\ x_2 \\ x_3 \end{array}\right) \mapsto \left(\begin{array}{c} x_1 \\ x_2 \end{array}\right)$

の表現行列を求めます．

まず，
${\bf v}_1 = \left(\begin{array}{c} 1 \\ 0 \\ 0 \end{array}\right)$, 
${\bf v}_2 = \left(\begin{array}{c} 0 \\ 1 \\ 0 \end{array}\right)$, 
${\bf v}_3 = \left(\begin{array}{c} 0 \\ 0 \\ 1 \end{array}\right)$
とし，
${\bf w}_1 = \left(\begin{array}{c} 1 \\ 0 \end{array}\right)$, 
${\bf w}_2 = \left(\begin{array}{c} 0 \\ 1 \end{array}\right)$
とします．
これらがそれぞれ$\mathbb{R}^3$と$\mathbb{R}^2$の基底になっていることは明らかです．

射影の定義より，次の式が成り立ちます．

$P_{12}({\bf v}_1) = {\bf w}_1$, <br>
$P_{12}({\bf v}_2) = {\bf w}_2$, <br>
$P_{12}({\bf v}_3) = {\bf 0}$

これをまとめて行列にすると，

$\left(\begin{array}{c} P_{12}({\bf v}_1) & P_{12}({\bf v}_2) & P_{12}({\bf v}_3) \end{array}\right) = 
\left(\begin{array}{c} {\bf w}_1 & {\bf w}_2 & {\bf 0} \end{array}\right) =
\left(\begin{array}{c} {\bf w}_1 & {\bf w}_2 \end{array}\right) 
\left(\begin{array}{c} 1 & 0 & 0 \\ 0 & 1 & 0 \end{array}\right) $

となります．
したがって，$\mathbb{R}^3$の第1軸と第2軸への射影$P_{12}$の表現行列は次のものになります．

${\bf P}_{12} = \left(\begin{array}{c} 1 & 0 & 0 \\ 0 & 1 & 0 \end{array}\right)$

同様にして，第1軸と第3軸への射影$P_{13}$や第2軸と第3軸への射影$P_{23}$も求められます．

${\bf P}_{13} = \left(\begin{array}{c} 1 & 0 & 0 \\ 0 & 0 & 1 \end{array}\right)$, 
${\bf P}_{23} = \left(\begin{array}{c} 0 & 1 & 0 \\ 0 & 0 & 1 \end{array}\right)$

## 多項式の微分の表現行列

ここでは読み易さを優先して，3次多項式の微分について説明します．

3次多項式のベクトル空間は，
$\mathscr{P}_3 = \left\{P_3(x) \mid P_3(x) = p_0 + p_1x + p_2x^2 + p_3x^3 ,\, p_i \in K \right\}$
です．
このベクトル空間は4次元です．

3次多項式の微分は，

> 
$ \frac{d}{dx} P_3(x) =
\frac{d}{dx} (p_0 + p_1x + p_2x^2 + p_3x^3) =
p_1 + 2p_2x + 3p_3x^2 =
P'_2(x) $

ですので，$\mathscr{P}_3$から$\mathscr{P}_2$への線形写像です．

$\mathscr{P}_3$の基底を，
${\bf v}_1 = 1$, 
${\bf v}_2 = x$, 
${\bf v}_3 = x^2$, 
${\bf v}_4 = x^3$
<br>
$\mathscr{P}_2$の基底を，
${\bf w}_1 = 1$, 
${\bf w}_2 = x$, 
${\bf w}_3 = x^2$

とします．

$\mathscr{P}_3$の基底に微分$\frac{d}{dx}$を作用させると，

$\frac{d}{dx}{\bf v}_1 = \frac{d}{dx}1 = {\bf 0}$, <br>
$\frac{d}{dx}{\bf v}_2 = \frac{d}{dx}x = 1 = {\bf w}_1$, <br>
$\frac{d}{dx}{\bf v}_3 = \frac{d}{dx}x^2 = 2x = 2{\bf w}_2$, <br>
$\frac{d}{dx}{\bf v}_4 = \frac{d}{dx}x^3 = 3x^2 = 3{\bf w}_3$

となります．
これをまとめて行列にすると，

$\left(\begin{array}{c} \frac{d}{dx}{\bf v}_1 & \frac{d}{dx}{\bf v}_2 & \frac{d}{dx}{\bf v}_3 & \frac{d}{dx}{\bf v}_4 \end{array}\right) = 
\left(\begin{array}{c} 0 & {\bf w}_1 & 2{\bf w}_2 & 3{\bf w}_3  \end{array}\right) =
\left(\begin{array}{c} {\bf w}_1 & {\bf w}_2 & {\bf w}_3 \end{array}\right) 
\left(\begin{array}{c} 0 & 1 & 0 & 0 \\ 
                       0 & 0 & 2 & 0 \\
                       0 & 0 & 0 & 3 
                       \end{array}\right) $

となります．
したがって，3次多項式から2次多項式への線形写像としての微分は次の行列で表現されます．

$ \frac{d}{dx} =
\left(\begin{array}{c} 
0 & 1 & 0 & 0 \\
0 & 0 & 2 & 0 \\
0 & 0 & 0 & 3 
\end{array}\right) $

## 拡大・縮小の表現行列

2次元の実数ベクトル${\bf x} = \left(\begin{array}{c} x_1 \\ x_2 \end{array}\right)$に対して，
ゼロでない実数$\lambda \neq 0$によって拡大・縮小をする線形変換$E_{\lambda}$を行列で表します．

2次元の実数ベクトル空間$\mathbb{R}^2$の基底を
${\bf v}_1 = \left(\begin{array}{c} 1 \\ 0 \end{array}\right)$, 
${\bf v}_2 = \left(\begin{array}{c} 0 \\ 1 \end{array}\right)$
とします．

拡大・縮小の作用素$E_{\lambda}$を基底に作用させます．

$E_{\lambda}{\bf v}_1 = E_{\lambda}\left(\begin{array}{c} 1 \\ 0 \end{array}\right) = 
{\lambda}\left(\begin{array}{c} 1 \\ 0 \end{array}\right) = {\lambda}{\bf v}_1$, 
<br>
$E_{\lambda}{\bf v}_2 = E_{\lambda}\left(\begin{array}{c} 0 \\ 1 \end{array}\right) = 
{\lambda}\left(\begin{array}{c} 0 \\ 1 \end{array}\right) = {\lambda}{\bf v}_2$

となります．
これをまとめて行列で表現すると，

$\left(\begin{array}{c} E_{\lambda}{\bf v}_1 & E_{\lambda}{\bf v}_2 \end{array}\right) = 
\left(\begin{array}{c} {\lambda}{\bf v}_1 & {\lambda}{\bf v}_2 \end{array}\right) =
\left(\begin{array}{c} {\bf v}_1 & {\bf v}_2 \end{array}\right) 
\left(\begin{array}{c} \lambda & 0  \\ 0 & \lambda \end{array}\right) $

となります．
したがって，2次元の実数ベクトル空間上の拡大・縮小の線形変換は次の行列となります．

$E_{\lambda} = \left(\begin{array}{c} \lambda & 0  \\ 0 & \lambda \end{array}\right)$

## 回転の表現行列

2次元の実数ベクトル空間上の回転は線形変換ですが，これを行列で表すには三角関数の力を借りる必要があります．

2次元の実数ベクトル空間$\mathbb{R}^2$の基底を
${\bf v}_1 = \left(\begin{array}{c} 1 \\ 0 \end{array}\right)$, 
${\bf v}_2 = \left(\begin{array}{c} 0 \\ 1 \end{array}\right)$
としたときに，この基底ベクトルを反時計回りに $\theta$度回転 させたベクトルを下図に示します．

![2次元の回転](./images/linearmap011.jpg)

これにより，基底の回転は，

$R_{\theta}{\bf v}_1 = R_{\theta}\left(\begin{array}{c} 1 \\ 0 \end{array}\right) = 
\left(\begin{array}{c} \cos\theta \\ \sin\theta \end{array}\right) = 
\cos\theta\left(\begin{array}{c} 1 \\ 0 \end{array}\right) + \sin\theta\left(\begin{array}{c} 0 \\ 1 \end{array}\right) =
\cos\theta{\bf v}_1+\sin\theta{\bf v}_2$, 
<br>
$R_{\theta}{\bf v}_2 = R_{\theta}\left(\begin{array}{c} 0 \\ 1 \end{array}\right) = 
\left(\begin{array}{c} -\sin\theta \\ \cos\theta \end{array}\right) = 
-\sin\theta\left(\begin{array}{c} 1 \\ 0 \end{array}\right) + \cos\theta\left(\begin{array}{c} 0 \\ 1 \end{array}\right) =
-\sin\theta{\bf v}_1+\cos\theta{\bf v}_2$

となります．
これをまとめて，行列にすると，

$\left(\begin{array}{c} R_{\theta}{\bf v}_1 & R_{\theta}{\bf v}_2 \end{array}\right) = 
\left(\begin{array}{c} \cos\theta{\bf v}_1+\sin\theta{\bf v}_2 & -\sin\theta{\bf v}_1+\cos\theta{\bf v}_2 \end{array}\right) =
\left(\begin{array}{c} {\bf v}_1 & {\bf v}_2 \end{array}\right) 
\left(\begin{array}{c} \cos\theta & -\sin\theta  \\ \sin\theta & \cos\theta \end{array}\right) $

となります．
したがって，2次元の実数ベクトル空間上の回転の線形変換は次の行列となります．

$R_{\theta} = \left(\begin{array}{c} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{array}\right)$

### 三角関数の加法定理

回転の表現行列を利用すると，三角関数の加法定理が簡単に求まります．
回転の合成には，次の等式が成り立ちました．

> $R_{\eta} \circ R_{\theta} = R_{\eta+\theta}$

この式に，

$R_{\eta} = \left(\begin{array}{c} \cos\eta & -\sin\eta \\ \sin\eta & \cos\eta \end{array}\right)$, <br>
$R_{\theta} = \left(\begin{array}{c} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{array}\right)$, <br>
$R_{\eta+\theta} = 
\left(\begin{array}{c} \cos(\eta+\theta) & -\sin(\eta+\theta) \\ \sin(\eta+\theta) & \cos(\eta+\theta) \end{array}\right)$

を代入して，行列の積の計算を行います．

$ \left(\begin{array}{c} \cos(\eta+\theta) & -\sin(\eta+\theta) \\ \sin(\eta+\theta) & \cos(\eta+\theta) \end{array}\right) =
R_{\eta+\theta} = 
R_{\eta} \circ R_{\theta} = 
\left(\begin{array}{c} \cos\eta & -\sin\eta \\ \sin\eta & \cos\eta \end{array}\right)
\left(\begin{array}{c} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{array}\right)
=
\left(\begin{array}{c} 
    \cos\eta\cos\theta-\sin\eta\sin\theta & -\cos\eta\sin\theta-\sin\eta\cos\theta \\
    \sin\eta\cos\theta+\cos\eta\sin\theta & -\sin\eta\sin\theta+\cos\eta\cos\theta 
\end{array}\right) $

最左辺の行列と最右辺の行列の成分どうしが一致しているので，次に示す三角関数の加法定理の式を得ることができます．

> $\cos(\eta+\theta) = \cos\eta\cos\theta-\sin\eta\sin\theta$

> $\sin(\eta+\theta) = \sin\eta\cos\theta+\cos\eta\sin\theta$

線形代数を学ぶことによって，三角関数の定理が導かれるところが素晴らしいです．

*****
線形写像の表現行列の定理のおかげで，線形代数の構造を調べるには行列を研究すれば良いことが分かりました．
