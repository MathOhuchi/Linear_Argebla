# 群（Group）

代数で扱う最も基本的な構造が群（Group）です．
私達は群について本格的な学習は行いませんが，線形代数の中でもしばしば群構造が現れますので，群が何であるかを知っておく必要はあります．
群そのものは単純な構造であるために，定義について理解することは易しく思えるでしょう．
しかし，その活用範囲は数学全体に及んでいて，それらの理論の中に群構造を発見することが出来るでしょう．


## 群の定義

集合$G$に二項演算 $\cdot$ が定義されていて，任意の$a,b,c \in G$について，以下の条件を満たすとき，集合$G$を群と言います．

- <font color=blue>結合法則</font> $(a{\cdot}b){\cdot}c = a{\cdot}(b{\cdot}c)$が成立する．
- $a{\cdot}e = e{\cdot}a = a$ となる<font color=blue>単位元</font> $e \in G$が存在する．
- 任意の$a \in G$に対して，$ab = ba = e$となる<font color=blue>逆元</font> $b$ が存在する．この逆元$b$を $a^{-1}$と表します．


## 置換

識別可能な$n$個のものに所有者がいる状況で，その所有者が替わることを使って置換を説明します．

分かりやすい例として，A,B,Cの3人が赤青緑の玉を持っています．
最初は，Aが赤，Bが青，Cが緑を持っています．
これを入替えて，Aが青，Bが緑，Cが赤を持ちました．

![色の置換](./images/group001.jpg)

ここで人の並び順はは気にせず，「赤玉を青玉に替えた」，「青玉を緑玉に替えた」，「緑玉を赤玉に替えた」という事実に注目します．
この玉の置き換えを<font color=blue>置換</font>と言います．
この置換の個別の要素の関係を関数$\sigma(\cdot)$を使って，

> $\sigma(\text{赤}) = \text{青}$ <br>
> $\sigma(\text{青}) = \text{緑}$ <br>
> $\sigma(\text{緑}) = \text{赤}$

と表します．
図から分かるように，置換前と後で，1対1の対応が付いています．
すなわち，「$\sigma(\text{赤}) = \text{青}$かつ $\sigma(\text{緑}) = \text{青}$」というような状況は起こりません．

さて，色を使った説明は分かりやすいのですが，一般化するには不便です．そこで色の代わりに数字を使うことにします．
ここでは，赤を1，青を2，緑を3とします．すると，トレイの玉の図は次のようになります．

![数字の置換](./images/group002.jpg)

さらに個別要素の関係を表す関数も次のようになります．

> $\sigma(1) = 2$ <br>
> $\sigma(2) = 3$ <br>
> $\sigma(3) = 1$

そして，この置換の個別関数をまとめて$\sigma$として，次のように表します．

> 
$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
2 & 3 & 1 \\
\end{array}\right) $

ただし，この表現はここだけの特殊な記法です．
一般の数学書では
$ \left(\begin{array}{c}
1 & 2 & 3 \\
2 & 3 & 1 \\
\end{array}\right) $
と記載されます．
しかし，この記法は行列の記載法と混同しやすいので敢えて冗長な表現を用いました．


### 同じ置換

次の2つの置換を比べてみましょう．

$ \sigma =
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
2 & 3 & 1 \\
\end{array}\right) $
, 
$ \sigma' =
\left(\begin{array}{c}
3 & 1 & 2 \\
\downarrow & \downarrow &\downarrow \\
1 & 2 & 3 \\
\end{array}\right) $

一見，異なる置換のように見えますが，実は同じ置換を表しています．
$\sigma$は，上で示した置換ですが，同じように$\sigma'$についても図で表現してみましょう．

![数字の置換2](./images/group003.jpg)

これは，ABCの3人の場所が変わっただけで，
$\sigma(1) = 2$, 
$\sigma(2) = 3$, 
$\sigma(3) = 1$
という置き換えは変わりません．
つまり，この2つの置換$\sigma$と$\tau$は同じ置換を表しています．

さらに一歩踏み込んで，次の状況を考えます．

![数字の置換3](./images/group004.jpg)

このケースは持っている人が入れ替わっています．人と玉との組み合わせを考えると，$\sigma$と異なります．
しかし，ここでは「持っている人は誰か」という情報を除外して考えます．
すると，残った情報は，「<font color=blue>1を2に替えた</font>」，「<font color=blue>2を3に替えた</font>」，「<font color=blue>3を1に替えた</font>」という事だけです．

結局，「（前）が（後）に替わった」という情報だけで同じ置換か否かを判断します．
ここで扱った$\sigma$と同じ置換を列記します．

$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
2 & 3 & 1 \\
\end{array}\right) $
$=$
$ \left(\begin{array}{c}
1 & 3 & 2 \\
\downarrow & \downarrow &\downarrow \\
2 & 1 & 3 \\
\end{array}\right) $
$=$
$ \left(\begin{array}{c}
2 & 1 & 3 \\
\downarrow & \downarrow &\downarrow \\
3 & 2 & 1 \\
\end{array}\right) $
$=$
$ \left(\begin{array}{c}
2 & 3 & 1 \\
\downarrow & \downarrow &\downarrow \\
3 & 1 & 2 \\
\end{array}\right) $
$=$
$ \left(\begin{array}{c}
3 & 1 & 2 \\
\downarrow & \downarrow &\downarrow \\
1 & 2 & 3 \\
\end{array}\right) $
$=$
$ \left(\begin{array}{c}
3 & 2 & 1 \\
\downarrow & \downarrow &\downarrow \\
1 & 3 & 2 \\
\end{array}\right) $

このように，3個の集合の置換を表す表現では1つの置換に必ず6通りの表現があります．
この6通りの表現による置換は同じなので，表記方法が異なっても同じ置換と見なします．
そこで，置換前の数字が左から1,2,3 と並んでいる置換を標準的な表記とします．

### 対称群

3個の要素の集合$\{1,2,3\}$の全ての置換を列記します．

$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
1 & 2 & 3 \\
\end{array}\right) $
, 
$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
1 & 3 & 2 \\
\end{array}\right) $
, 
$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
2 & 1 & 3 \\
\end{array}\right) $
, 
$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
2 & 3 & 1 \\
\end{array}\right) $
, 
$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
3 & 1 & 2 \\
\end{array}\right) $
, 
$ \left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
3 & 2 & 1 \\
\end{array}\right) $



本質的の異なる置換はこの6種類です．そして，それぞれの置換について6通りの表記方法があるということです．
この6個の置換を要素とする集合を$S_3$と記載します．
この集合は置換を続けて行うことを積として群になっています．
この$S_3$を<font color=blue>3次対称群</font>と言います．

以下で，群であることを照明します．

#### 置換の積

$\sigma = 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\sigma(1) & \sigma(2) & \sigma(3) \\
\end{array}\right) $
, 
$\tau = 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\tau(1) & \tau(2) & \tau(3) \\
\end{array}\right) $

としたとき，この合成$\sigma\tau$を次のように定めます．

$\sigma \tau = 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\tau(\sigma(1)) & \tau(\sigma(2)) & \tau(\sigma(3)) \\
\end{array}\right) $

具体的な計算例を一つ行います．

$ 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
3 & 1 & 2 \\
\end{array}\right) 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
1 & 3 & 2 \\
\end{array}\right)
=
\begin{array}{c}
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
3 & 1 & 2 \\
\end{array}\right) 
\\
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
1 & 3 & 2 \\
\end{array}\right)
\end{array}
=
\begin{array}{c}
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
3 & 1 & 2 \\
\end{array}\right) 
\\
\left(\begin{array}{c}
3 & 1 & 2 \\
\downarrow & \downarrow &\downarrow \\
2 & 1 & 3 \\
\end{array}\right)
\end{array}
=
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
3 & 1 & 2 \\
\downarrow & \downarrow &\downarrow \\
2 & 1 & 3 \\
\end{array}\right)
=
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
2 & 1 & 3 \\
\end{array}\right)
$


#### 結合法則

$S_3$の3つの要素$\sigma$, $\tau$, $\upsilon$を次のように定義します．

$\sigma = 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\sigma(1) & \sigma(2) & \sigma(3) \\
\end{array}\right) $
, 
$\tau = 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\tau(1) & \tau(2) & \tau(3) \\
\end{array}\right) $
, 
$\upsilon = 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\upsilon(1) & \upsilon(2) & \upsilon(3) \\
\end{array}\right) $
とします．

このとき，

$(\sigma\tau)\upsilon =
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\tau(\sigma(1)) & \tau(\sigma(2)) & \tau(\sigma(3)) \\
\end{array}\right)
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\upsilon(1) & \upsilon(2) & \upsilon(3) \\
\end{array}\right)
=
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\upsilon(\tau(\sigma(1))) & \upsilon(\tau(\sigma(2))) & \upsilon(\tau(\sigma(3))) \\
\end{array}\right) $

かつ

$\sigma(\tau\upsilon) =
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\sigma(1) & \sigma(2) & \sigma(3) \\
\end{array}\right)
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\upsilon(\tau(1)) & \upsilon(\tau(2)) & \upsilon(\tau(3)) \\
\end{array}\right)
=
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\upsilon(\tau(\sigma(1))) & \upsilon(\tau(\sigma(2))) & \upsilon(\tau(\sigma(3))) \\
\end{array}\right) $

となるので，$(\sigma\tau)\upsilon = \sigma(\tau\upsilon)$が成り立ちます．

#### 単位元

$S_3$において，何も置換しないことが単位元になります．

> 
$ e =
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
1 & 2 & 3 \\
\end{array}\right) $

#### 逆元

$S_3$の任の要素に対して，前と後を入替えたものが逆元となります．

$\sigma = 
\left(\begin{array}{c}
1 & 2 & 3 \\
\downarrow & \downarrow &\downarrow \\
\sigma(1) & \sigma(2) & \sigma(3) \\
\end{array}\right) $
の逆元は，
$\sigma^{-1} = 
\left(\begin{array}{c}
\sigma(1) & \sigma(2) & \sigma(3) \\
\downarrow & \downarrow &\downarrow \\
1 & 2 & 3 \\
\end{array}\right) $
となります．

ここでは，3次対称群について説明しましたが，一般化して$n$個の要素からなる集合も置換の連結を積として群になります．
これを<font color=blue>$n$次対称群</font>と言います．

*****
### 積の概念

群において二項演算の積が中心的役割を果たしますが，積という呼び方が数値の掛け算を強く印象づけられます．
しかし，対称群の例にもあるように，<font color=blue>積は変換の連結</font>というイメージを持つことが，理論を深めていく上で適切です．
